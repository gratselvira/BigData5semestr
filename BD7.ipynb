```python
# Задание 7.1. Напишите программу, которая реализует In-mapper combining v.1 для задачи WordCount.
class MapperV1:
    def Map(self, doc):
        H = {}     
        for word in doc.split():
            if word in H:
                H[word] += 1
            else:
                H[word] = 1       
        for word, count in H.items(): #вывод
            print(f"{word}\t{count}")


if __name__ == "__main__":
    with open("text7.txt", "r") as file:
        text = file.read()

    mapper = MapperV1() # Создание экземпляра класса
    for line in text.splitlines():
        mapper.Map(line)
```

```python
# Задание 7.2. Напишите программу, которая реализует In-mapper combining v.2 для задачи WordCount.
class MapperV2:
    def __init__(self):
        self.H = {}

    def Map(self, doc):
        for word in doc.split():
            if word in self.H:
                self.H[word] += 1
            else:
                self.H[word] = 1

    def Close(self):
        for word, count in self.H.items():
            print(f"{word}\t{count}")

if __name__ == "__main__":
    mapper = MapperV2()    
    with open("text7.txt", "r") as file:
        text = file.read()
    for line in text.splitlines():
        mapper.Map(line)
    mapper.Close()
```

```python
# Задание 7.3. Реализуйте reducer в задаче подсчета среднего времени, проведенного пользователем на странице.

from collections import defaultdict
class AverageTimeReducer:
    def __init__(self):
        self.page_data = defaultdict(lambda: [0, 0])  # Словарь с начальным значением [сумма, количество]

    def process_data(self, mapper_output):
        for page, time in mapper_output:
            self.page_data[page][0] += time  # Суммируем время
            self.page_data[page][1] += 1    # Увеличиваем счетчик записей

    def calculate_average(self):
        averages = {}
        for page, (total_time, count) in self.page_data.items():
            averages[page] = total_time // count  # Среднее время (округление до целого числа)
        return averages

    def display_result(self):
        averages = self.calculate_average()
        for page, avg_time in averages.items():
            print(f"{page} {avg_time}")


# Основная функция
def main():
    # Данные, которые обычно поступают от Mapper
    mapper_output = [
        ("www.facebook.com", 100),
        ("www.google.com", 10),
        ("www.mail.ru", 20),
        ("www.yandex.ru", 40),
        ("www.google.com", 5),
        ("www.mail.ru", 50),
        ("www.google.com", 15),
        ("www.stepic.org", 60),
        ("www.yandex.ru", 30),
        ("www.stepic.org", 100),
        ("www.mail.ru", 60),
        ("www.facebook.com", 100),
        ("www.google.com", 10),
        ("www.stepic.org", 80),
        ("www.yandex.ru", 50),
        ("www.google.com", 20),
        ("www.yandex.ru", 70),
        ("www.mail.ru", 40),
    ]

    reducer = AverageTimeReducer()
    reducer.process_data(mapper_output)
    reducer.display_result()


main()
```

```python
# Задание 7.4. Реализуйте Combiner в задаче подсчета среднего времени, проведенного пользователем на странице.
from collections import defaultdict

class TimeCombiner:
    def __init__(self):
        self.page_data = defaultdict(lambda: [0, 0])  

    def process_data(self, mapper_output):
        for record in mapper_output:
            page, value = record.split()  # Разделение строки на ключ и значение
            time, count = map(int, value.split(";"))  # Разделение "время;1" на числа
            self.page_data[page][0] += time  # Суммируем время
            self.page_data[page][1] += count  # Увеличиваем количество записей

    def get_combined_output(self):
        combined_data = []
        for page, (total_time, count) in self.page_data.items():
            combined_data.append(f"{page} {total_time};{count}")
        return combined_data

    def display_result(self):
        combined_data = self.get_combined_output()
        for record in combined_data:
            print(record)


# Основная функция
def main():
    # Входные данные от Mapper
    mapper_output = [
        "www.facebook.com 100;1",
        "www.google.com 10;1",
        "www.mail.ru 20;1",
        "www.yandex.ru 40;1",
        "www.google.com 5;1",
        "www.mail.ru 50;1",
        "www.google.com 15;1",
        "www.stepic.org 60;1",
        "www.yandex.ru 30;1",
        "www.stepic.org 100;1",
        "www.mail.ru 60;1",
        "www.facebook.com 100;1",
        "www.google.com 10;1",
        "www.stepic.org 80;1",
        "www.yandex.ru 50;1",
        "www.google.com 20;1",
        "www.yandex.ru 70;1",
        "www.mail.ru 40;1"
    ]

    combiner = TimeCombiner()
    combiner.process_data(mapper_output)
    combiner.display_result()


main()
```
