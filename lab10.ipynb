```python
import re

file_path = "articles-sample.txt"
with open(file_path, 'r', encoding='utf-8') as f:
    articles = [line.strip() for line in f.readlines()]

# Функция mapper
def mapper(articles):
    for article in articles:
        parts = article.split(" ", 1)  # Разделяем только на две части
        if len(parts) < 2:
            continue  # Если нет текста, пропускаем

        article_id = parts[0]  # Номер статьи
        text = parts[1]  # Текст статьи

        # Очищаем текст от знаков пунктуации и разбиваем на слова
        words = re.findall(r'\b[a-zA-Zа-яА-ЯёЁ]+\b', text.lower())  # Учитываем только буквы

        for word in words:
            yield word, article_id  # Возвращаем пару (слово, id статьи)

# Функция reducer
def reducer(mapped_data):
    word_to_first_article = {}

    for word, article_id in mapped_data:
        if word not in word_to_first_article:  # Если слово еще не встречалось
            word_to_first_article[word] = article_id  # Сохраняем ID первой статьи

    # Сортируем результат по алфавиту
    sorted_result = sorted(word_to_first_article.items())

    return sorted_result

# Запуск mapper и reducer
mapped_data = list(map(mapper, [articles]))  # Запускаем маппер
flattened_mapped_data = [item for sublist in mapped_data for item in sublist]  # Разворачиваем список
final_result = reducer(flattened_mapped_data)  # Запускаем редьюсер

# Вывод результатов
for term, article_id in final_result:
    print(f"{term}: {article_id}")
```

```python
import re

file_path = "articles-sample.txt"
with open(file_path, 'r', encoding='utf-8') as f:
    articles = [line.strip() for line in f.readlines()]
    
def mapper(articles):
    for article in articles:
        parts = article.split(" ", 1)  # Разделяем только на две части
        if len(parts) < 2:
            continue  # Если нет текста, пропускаем

        article_id = parts[0]  # Номер статьи
        text = parts[1]  # Текст статьи

        # Очищаем текст от знаков пунктуации и разбиваем на слова
        words = re.findall(r'\b[a-zA-Zа-яА-ЯёЁ]+\b', text.lower())  # Учитываем только буквы, без цифр

        for word in words:
            yield word, article_id  # Возвращаем пару (слово, id статьи)

# Функция reducer
def reducer(mapped_data):
    word_info = {}

    for word, article_id in mapped_data:
        if word not in word_info:
            word_info[word] = {'count': 0, 'article_ids': []}  # Инициализируем словарь для слова

        word_info[word]['count'] += 1  # Увеличиваем счетчик вхождений
        if article_id not in word_info[word]['article_ids']:
            word_info[word]['article_ids'].append(article_id)  # Добавляем ID статьи, если его еще нет
    # Сортируем результат по алфавиту
    sorted_result = sorted(word_info.items())
    return sorted_result

# Запуск mapper и reducer
mapped_data = list(map(mapper, [articles]))  # Запускаем маппер
flattened_mapped_data = [item for sublist in mapped_data for item in sublist]  # Разворачиваем список
final_result = reducer(flattened_mapped_data)  # Запускаем редьюсер

# Вывод результатов
for term, info in final_result:
    print(f"{term}: {info['count']} [{', '.join(info['article_ids'])}]")
```
